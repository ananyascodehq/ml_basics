1. IMPORT LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

2. IMPORT THE DATASET

```python
dataset = pd.read_csv('Data.csv')

#matrix of features
X = dataset.iloc[:, :-1].values #all rows, all coloumns except -1 (last col)

#dependant variable
y = dataset.iloc[:, -1].values
```

```python
print(X)

print(y)
```

3. HANDLE MISSING VALUES

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

#connect the dataset to the imputer by using the fit method
imputer.fit(X[:, 1:3]) #only and salary coloumns

#general rule: apply the imputer to all numeric coloumns

#transform the dataset
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```python
print(X)
```

4. ENCODING CATEGORICAL DATA

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
# passthrough = keep the coloumns that does not go through the encoding process,
# in this case, age and salary


# the return type of ct.fit_transform(X) is not an ndarray
# but we need X as an ndarray to train our model so we convert using np.array() method
X = np.array(ct.fit_transform(X))
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(X)
print(y)
```

5. SPLITTING THE DATASET INTO TRAIN AND TEST SET

```python
from sklearn.model_selection import train_test_split

# test_size = 0.2 implies 20% observations in the test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

6. FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])

#important

# why X_test = sc.fit_transform(X_test) is wrong? fit_transform would
# compute a new mean, compute a new standard deviation, That creates a different scaler

# X_test represents new, unseen data (future / production data).

#correct approach:
X_test[:, 3:] = sc.transform(X_test[:, 3:])
```
